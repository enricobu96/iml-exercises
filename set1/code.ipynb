```python
import numpy as np
import pandas as pd

def create_dataset(n):
    x = np.random.uniform(low=-3, high=3, size=n)
    eps = np.random.normal(loc=0, scale=0.4, size=n)
    y = []
    for i in range(0, len(x)):
        y.append((1+x[i]-(x[i]**2)/2) + eps[i])
    d = {'x':x,'y':y}
    return pd.DataFrame(d)

training = create_dataset(20)
validation = create_dataset(20)
test = create_dataset(1000)
```

```python
# task b
def transform_x(x, degree):
    X = []
    for xi in x:
        row = [xi**d for d in range(0,degree+1)]
        X.append(row)
    return X

def ols(x, y, deg):
    X = np.matrix(transform_x(x, deg))
    betas = -np.matmul(np.matmul(np.linalg.inv(np.matmul(X.transpose(), X)), X.transpose()), y)
    return betas

def f(x, betas):
    y = betas[0]
    for i in range(1,len(betas)):
        y+=betas[i]*(x**i)
    return -y

degrees = range(0,11)
for degree in degrees:
    betas = ols(training['x'], training['y'], degree)
    line_x = np.linspace(start=-3, stop=3, num=256)
    line_y = np.array([f(xi, betas) for xi in line_x])
    plt.scatter(training['x'], training['y'])
    plt.plot(line_x, line_y)
    plt.show()

def calculate_mse(degree, betas, tr):
    pred = []
    for i in range(0, len(tr)):
        pred.append(f(tr['x'][i], betas))
    mse = np.sum((tr['y']-pred)**2)
    mse/=len(tr)
    return mse

mses = []
for degree in degrees:
    betas = ols(training['x'], training['y'], deg=degree)
    mses.append(calculate_mse(degree, betas, training))

d = {'deg': degrees, 'mse': mses}
result = pd.DataFrame(d)
result
```

```python
print('MSEs on validation set')
mses = []
for degree in degrees:
    betas = ols(training['x'], training['y'], deg=degree)
    mses.append(calculate_mse(degree, betas, validation))

d = {'deg': degrees, 'mse': mses}
result = pd.DataFrame(d)
display(result)

print('MSEs on test set')
mses = []
for degree in degrees:
    betas = ols(training['x'], training['y'], deg=degree)
    mses.append(calculate_mse(degree, betas, test))

d = {'deg': degrees, 'mse': mses}
result = pd.DataFrame(d)
display(result)
```

```python
from copy import deepcopy

tv_combined = training.append(validation).reset_index()

def cv_split(tosplit, folds):
    split = []
    copy = deepcopy(tosplit).to_numpy()
    for i in range(0, folds):
        t = []
        while len(t) < int(len(tosplit) / folds):
            toremove = np.random.randint(0, len(copy))
            t.append([copy[toremove][1], copy[toremove][2]])
            copy = np.delete(copy, toremove, 0)
        split.append(t)
    return split

splits = cv_split(tv_combined, 10)
for degree in range(0,11):
    cv = 0
    for i in range(0, 10):
        validat_set = pd.DataFrame(splits[i], columns=['x','y'])
        train_set = pd.DataFrame(columns=['x','y'])
        for j in range(0,10):
            if(j!=i):
                sp = pd.DataFrame(splits[j], columns=['x','y'])
                train_set = train_set.append(sp, ignore_index=True)
        
        min_mse = float('inf')
        betas = ols(training['x'], training['y'], deg=degree)
        mse = calculate_mse(degree, betas, train_set)
        cv+=mse
        if mse < min_mse:
            min_mse = mse
            min_mse_index = i
    cv/=10
    print('At degree', degree, 'found minimum MSE',min_mse, 'at index', min_mse_index)
    print('At this degree, cv is', cv)


```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.model_selection import KFold

bcnp = pd.read_csv('Bias_correction_ucl.csv')
bcnp = bcnp.drop(['station', 'Date', 'Next_Tmin'], axis=1)
bcnp = bcnp.dropna(how='any', axis=0)

training_set = bcnp.sample(n=100)
test_set = pd.concat([bcnp, training_set]).drop_duplicates(keep=False)

# linear regression
reg = LinearRegression().fit(training_set[training_set.columns.difference(['Next_Tmax'])], training_set['Next_Tmax'])
y_pred = reg.predict(test_set[test_set.columns.difference(['Next_Tmax'])])
rmse = mean_squared_error(test_set['Next_Tmax'], y_pred)
print('For linear regression, RMSE on test is', np.sqrt(rmse))

y_pred_t = reg.predict(training_set[test_set.columns.difference(['Next_Tmax'])])
rmse_train = mean_squared_error(training_set['Next_Tmax'], y_pred_t)
print('rmse on train', np.sqrt(rmse_train))

# regression tree
reg = DecisionTreeRegressor(random_state=0).fit(training_set[training_set.columns.difference(['Next_Tmax'])], training_set['Next_Tmax'])
y_pred = reg.predict(test_set[test_set.columns.difference(['Next_Tmax'])])
rmse = mean_squared_error(test_set['Next_Tmax'], y_pred)
print('For regression tree, RMSE on test is', np.sqrt(rmse))
y_pred_t = reg.predict(training_set[test_set.columns.difference(['Next_Tmax'])])
rmse_train = mean_squared_error(training_set['Next_Tmax'], y_pred_t)
print('rmse on train', np.sqrt(rmse_train))

# random forest
reg = RandomForestRegressor(random_state=0).fit(training_set[training_set.columns.difference(['Next_Tmax'])], training_set['Next_Tmax'])
y_pred = reg.predict(test_set[test_set.columns.difference(['Next_Tmax'])])
rmse = mean_squared_error(test_set['Next_Tmax'], y_pred)
print('For random forest, RMSE on test is', np.sqrt(rmse))
y_pred_t = reg.predict(training_set[test_set.columns.difference(['Next_Tmax'])])
rmse_train = mean_squared_error(training_set['Next_Tmax'], y_pred_t)
print('rmse on train', np.sqrt(rmse_train))

# svm
reg = SVR().fit(training_set[training_set.columns.difference(['Next_Tmax'])], training_set['Next_Tmax'])
y_pred = reg.predict(test_set[test_set.columns.difference(['Next_Tmax'])])
rmse = mean_squared_error(test_set['Next_Tmax'], y_pred)
print('For SVM, RMSE on test is', np.sqrt(rmse))
y_pred_t = reg.predict(training_set[test_set.columns.difference(['Next_Tmax'])])
rmse_train = mean_squared_error(training_set['Next_Tmax'], y_pred_t)
print('rmse on train', np.sqrt(rmse_train))

# ridge
reg = Ridge().fit(training_set[training_set.columns.difference(['Next_Tmax'])], training_set['Next_Tmax'])
y_pred = reg.predict(test_set[test_set.columns.difference(['Next_Tmax'])])
rmse = mean_squared_error(test_set['Next_Tmax'], y_pred)
print('For Ridge, RMSE on test is', np.sqrt(rmse))
y_pred_t = reg.predict(training_set[test_set.columns.difference(['Next_Tmax'])])
rmse_train = mean_squared_error(training_set['Next_Tmax'], y_pred_t)
print('rmse on train', np.sqrt(rmse_train))

# cv losses

kf = KFold(n_splits=10)

## regression
cv = 0
for tr, val in kf.split(training_set):
    reg = LinearRegression().fit(training_set.iloc[tr][training_set.columns.difference(['Next_Tmax'])], training_set.iloc[tr]['Next_Tmax'])
    y_pred = reg.predict(training_set.iloc[val][training_set.columns.difference(['Next_Tmax'])])
    mse = mean_squared_error(training_set.iloc[val]['Next_Tmax'], y_pred)
    cv+=mse
cv/=10
print('For regression, cv loss is', cv)

## regression tree
cv = 0
for tr, val in kf.split(training_set):
    reg = DecisionTreeRegressor(random_state=0).fit(training_set.iloc[tr][training_set.columns.difference(['Next_Tmax'])], training_set.iloc[tr]['Next_Tmax'])
    y_pred = reg.predict(training_set.iloc[val][training_set.columns.difference(['Next_Tmax'])])
    mse = mean_squared_error(training_set.iloc[val]['Next_Tmax'], y_pred)
    cv+=mse
cv/=10
print('For regression tree, cv loss is', cv)

## random forest
cv = 0
for tr, val in kf.split(training_set):
    reg = RandomForestRegressor().fit(training_set.iloc[tr][training_set.columns.difference(['Next_Tmax'])], training_set.iloc[tr]['Next_Tmax'])
    y_pred = reg.predict(training_set.iloc[val][training_set.columns.difference(['Next_Tmax'])])
    mse = mean_squared_error(training_set.iloc[val]['Next_Tmax'], y_pred)
    cv+=mse
cv/=10
print('For random forest, cv loss is', cv)

## svm
cv = 0
for tr, val in kf.split(training_set):
    reg = SVR().fit(training_set.iloc[tr][training_set.columns.difference(['Next_Tmax'])], training_set.iloc[tr]['Next_Tmax'])
    y_pred = reg.predict(training_set.iloc[val][training_set.columns.difference(['Next_Tmax'])])
    mse = mean_squared_error(training_set.iloc[val]['Next_Tmax'], y_pred)
    cv+=mse
cv/=10
print('For svm, cv loss is', cv)

## ridge
cv = 0
for tr, val in kf.split(training_set):
    reg = Ridge().fit(training_set.iloc[tr][training_set.columns.difference(['Next_Tmax'])], training_set.iloc[tr]['Next_Tmax'])
    y_pred = reg.predict(training_set.iloc[val][training_set.columns.difference(['Next_Tmax'])])
    mse = mean_squared_error(training_set.iloc[val]['Next_Tmax'], y_pred)
    cv+=mse
cv/=10
print('For ridge, cv loss is', cv)
```
